```python
import pandas as pd
import numpy as np
from citipy import citipy
```

```python
lats= np.random.uniform(low=-90.00, high = 90.00 , size = 2000)
lngs= np.random.uniform(low = -180.00, high = 180.00, size = 2000)
lats_lngs = zip(lats,lngs)
lats_lngs
```

```python
coordinates =list(lats_lngs)
```

```python
cities=[]
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0],coordinate[1]).city_name
    if city not in cities:
        cities.append(city)
len(cities)
```

```python
import requests
from config import weather_api_key
```

```python
url = 'http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=' + weather_api_key
print(url)
```

```python
import time
from datetime import datetime
```

```python
city_data=[]
print('Beginning of data retrieval    ')
print('...............................')

record_count= 1
set_count = 1

for i, city in enumerate(cities):
    if (i % 50 ==0 and i>=50):
        set_count +=1
        record_count = 1
        time.sleep(60)
        
    city_url = url +'&q=' + cities[i]
    print(f'Processing Record{record_count} of set {set_count} | {city}')
    record_count += 1
    try:
        city_weather = requests.get(city_url).json()
        city_lat = city_weather['coord']['lat']
        city_lng = city_weather['coord']['lon']
        city_max_temp = city_weather['main']['temp_max']
        city_humidity = city_weather['clouds']['all']
        city_clouds=city_weather['wind']['speed']
        city_wind = city_weather['sys']['country']
        city_country = city_weather['sys']['country']
        city_weather_description = city_weather['weather'][0]['description']
        city_date = datetime.utcfromtimestamp(city_weather['dt']).strftime('%Y-%m-%d %H:%M%S')
        city_data.append({'City': city.title(),
                          'Lat': city_lat,
                          'Lng': city_lng,
                          'Max Temp' : city_max_temp,
                          'Humidity' : city_humidity,
                          'Cloudiness' : city_clouds,
                          'Wind Speed': city_wind,
                          'Country': city_country,
                          'Date': city_date,
                          'Current Description' : city_weather_description})
    except:
        print('City not found.Skipping...')
        pass
print('------------------------')
print('Data Retrival Complete')
print('------------------------')
```

```python
len(city_data)
```

```python
city_data_df = pd.DataFrame(city_data)
city_data_df
```

```python
new_coloumn_order=['City', 'Country', 'Lat','Lng' ,'Max Temp', 'Humidity', 'Cloudiness',
                   'Wind Speed', 'Current Description']
city_data_df=city_data_df[new_coloumn_order]
city_data_df.head(10)
```

```python
output_data_file= 'WeatherPy_Database.csv'
city_data_df.to_csv(output_data_file, index_label='City ID')
```
